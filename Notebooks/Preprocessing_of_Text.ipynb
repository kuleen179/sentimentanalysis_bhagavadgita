# Replace Archaic words

```python

for chapter in range(1, 19):
  with open('/content/drive/MyDrive/religious_texts/purohit swami/chapter wise text files/chapter ' + str(chapter) + '.txt') as f:

    content = f.read()

    temp = content.splitlines()

    res = []
    for line in temp:  
      for word in line.split():
        
        if word.isalnum() == False:
          # print(word)
          res2 = re.search(r'\W+', word).start()
          
          res3 = word[res2:len(word)]
          res2 = word[0 : res2]

          res.append(map_old_new.get(str(res2), str(res2)))

          res[len(res)-1] = res[len(res)-1] + res3

        else:
          # print(word)
          res.append(map_old_new.get(str(word), str(word)))

      res.append('\n')

    res = ' '.join(res)

    f.close()

    text_file = open('/content/drive/MyDrive/religious_texts/purohit swami/modified chapter wise text files/chapter ' + str(chapter) + '.txt', 'w')

    text_file.write(res)

    text_file.close()
```

# Convert verses to a single line

```python
for chapter in range (1, 19) 
  text_file = open('/content/drive/MyDrive/religious_texts/mahatma gandhi/chap ' + str(chapter) + '.txt')

  content = text_file.read()

  temp = content.splitlines()

  res = []

  string1 = ''

  for line in temp:  
    for chr in line:

      if chr.isdigit():
        if len(string1) > 3:
          res.append(string1)
          res.append('\n')
          string1 = ''

      elif chr == '\n':
        chr = ' ' 
        string1+=' '

      elif chr == '.' and len(string1)<3:
        continue

      else:
        string1+=chr

  res.append(string1)
  res = ' '.join(res)

  text_file = open('/content/drive/MyDrive/religious_texts/mahatma gandhi/chap ' + str(chapter) + ' modified.txt', 'w')

  text_file.write(res)

  text_file.close()
```

# Remove numbers and convert to dataframe

```python
for chapter in range(1, 19):
  filename = '/content/drive/MyDrive/religious_texts/purohit swami/modified chapter wise text files/chapter ' + str(chapter) + '.txt'
  
  with open(filename) as fh:
    content = fh.read()

    content_list = content.splitlines()
    content_list2 = []

    for line in content_list:
      remove_digits = str.maketrans('', '', digits)
      res = line.translate(remove_digits)

      content_list2.append(res)
        
    fh.close()
    
    print(len(content_list2))
  
  df = pd.DataFrame(content_list2, columns=['verse'])
  print(len(df))
  df.to_csv("/content/drive/MyDrive/religious_texts/purohit swami/csv folder/" + str(chapter) + ".csv")
```
